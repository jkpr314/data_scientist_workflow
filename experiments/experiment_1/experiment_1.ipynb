```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# load data
df = pd.read_parquet('../../data/country_year_gdp.parquet')
```

```python
countries = list(df.columns)
```

```python
country = 'LTU'
```

```python
df_growth = pd.DataFrame()
for country in countries:
    temp = df[country].dropna()
    temp = temp.to_frame()
    temp['GDP_shift'] = temp.shift()
    temp = temp.dropna()
    temp['growth'] = temp[country]/temp.GDP_shift - 1
    
    df_growth[country] = temp.growth
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.api import acf, graphics, pacf
from statsmodels.tsa.ar_model import AutoReg, ar_select_order
```

```python
df_growth.LTU.dropna()
```

```python
mod = AutoReg(df_growth.LTU.dropna().values,1, old_names=False)
res = mod.fit()
print(res.summary())
```
